```python
EOS = ["APR4_EPP", "BHF_BBB2", "H4", "HQC18", "KDE0V", "KDE0V1", "MPA1", "MS1_PP", "MS1B_PP", "RS", "SK255", "SK272", "SKI2", "SKI3", "SKI4", "SKI5", "SKI6", "SKMP", "SKOP", "SLy", "SLY2", "SLY9", "SLY230A"]
BayesFactor=[1.526, 1.555, 0.056, 1.422, 1.177, 1.283, 0.276, 0.001, 0.009, 0.176, 0.179, 0.159, 0.108, 0.107, 0.33, 0.025, 0.288, 0.29, 0.618, 1.0, 1.028, 0.37, 0.932]
```

```python
df = pd.read_csv(pathData+'real_data.csv')
columns = df[['m1_rec', 'm2_rec', 'chi1_rec', 'chi2_rec', 'snr']]
# Convert the DataFrame to a Numpy array
events = columns.values
pHasNS = np.zeros(len(df));  pHasREM = np.zeros(len(df));
```

```python
EOSdic = {}
count = 0
```

```python
for eos in EOS:
    print("Doing", eos)
    KNN = KNNreal.ClassificationKNN()
    KNN.loadModel(pathClassy+"optimal_models_final/", "knn_3cat_grid_eos_SLy")
    new_dic = {}
    new_dic['knn'] = KNN
    new_dic['bayes'] = BayesFactor[count]
    EOSdic[eos] = new_dic
    count = count + 1
    EOSdic[eos]['probs'] = KNN.model.predict_proba(events)
    pHasNS += (1 - EOSdic[eos]['probs'][:,0]) * EOSdic[eos]['bayes']
    pHasREM += EOSdic[eos]['probs'][:,2]  * EOSdic[eos]['bayes']
            
print('#'*60)

total = sum(BayesFactor)
```

```python
df['p(HasNS)_KNN'] = pHasNS/total
df['p(HasREM)_KNN'] = pHasREM/total

df.to_csv("real_data/final_table.csv")
```
