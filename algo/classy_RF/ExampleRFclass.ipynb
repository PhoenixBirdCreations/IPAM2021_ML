```python
RF = classyRF.ClassificationRF()
RF.load_dataset(path, fname_x='reg1_xtest.csv', fname_y='reg1_label.csv')
RF.split_train_test(pct=0.7)
```

```python
RF.train()
RF.print_metrics()
```

```python
# Prediction on a single event
m1  = 1.3
m2  = 1.2
pred_test = RF.compute_prediction([m1,m2])
print("Small binary")
print('input     :', np.array([[m1,m2]]))
print('prediction:', pred_test)
print('-'*60)

m1  = 2.1
m2  = 1.8
pred_test = RF.compute_prediction([m1,m2])
print("Big binary")
print('input     :', np.array([[m1,m2]]))
print('prediction:', pred_test)
print('-'*60)
```

```python
optimum = RF.crossvalidation()
RF.train( trees=optimum[0], criterion=optimum[1], max_features=optimum[2])
RF.print_metrics()
```
